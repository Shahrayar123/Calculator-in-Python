# Simple Calculator in Python
---

```python
def add(a,b):           
    return(a+b)

def sub(a,b):
    return(a-b)

def mul(a,b):
    return(a*b)

def div(a,b):
    try:
        return(a/b)
    except ZeroDivisionError:
        print("Zero Division Error arrise")
        
def mod(a,b):
    return(a%b)        

def power(a,b):
    return(a**b)

def mul(a,b):
    return(a*b)

while(True):
    print("\n------------------------>>> Enter your choice <<<------------------------\n")
    print("Enter 'A' for addition ")
    print("Enter 'S' for subtraction ")
    print("Enter 'M' for multiplication ")
    print("Enter 'D' for division ")
    print("Enter 'Mod' for modulus ")
    print("Enter 'P' for power ")
    print("Enter 'Q' for quit ")
    print()
    choice=input("Enter your choice: ")
    
    if(choice.upper()=='Q'):
        print("Thank you for using this calculator\nI hope you enjoy using this calculator \U0001F642 ")
        print("This calculator will be furthur improve in future")
        break
    
    
    elif(choice.upper()=='A'):
        num1=float(input("Enter first number: "))
        num2=float(input("Enter second number: "))
    
        result=add(num1,num2)
        print(f"After adding {num1} and {num2} the result is : {result}")
    
    elif(choice.upper()=='S'):
            
        num1=float(input("Enter first number: "))
        num2=float(input("Enter second number: "))
    
        result=sub(num1,num2)
        print(f"After subtraction {num1} and {num2} the result is : {result} ")
    
    elif(choice.upper()=='M'):
            
        num1=float(input("Enter first number: "))
        num2=float(input("Enter second number: "))
    
        result=mul(num1,num2)
        print(f"After multiplication {num1} and {num2} the result is : {result} ")
    
    elif(choice.upper()=='D'):
            
        num1=float(input("Enter first number: "))
        num2=float(input("Enter second number: "))
    
        result=div(num1,num2)
        if result==None:
            pass
        else:
            print(f"After dividing {num1} and {num2} the result is : {result} ")
        
    elif(choice.title()=='Mod'):
            
        num1=float(input("Enter first number: "))
        num2=float(input("Enter second number: "))
    
        result=mod(num1,num2)
        print(f"After taking modulus of {num1} and {num2} the result is : {result} ")
    
    elif(choice.upper()=='P'):
            
        num1=float(input("Enter first number: "))
        num2=float(input("Enter second number: "))
    
        result=power(num1,num2)
        print(f"After taking power of {num1}, {num2} times the result is : {result} ")
        
    else:
        print("Invalid choice ⚠️")
        

        
        
```
